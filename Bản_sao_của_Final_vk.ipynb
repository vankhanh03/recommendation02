```python
# Loại bỏ outliers
from typing import Tuple
from sklearn.base import BaseEstimator, TransformerMixin


def find_boxplot_boundaries(
    col: pd.Series, whisker_coeff: float = 1.5
) -> Tuple[float, float]:
    """Findx minimum and maximum in boxplot.

    Args:
        col: a pandas serires of input.
        whisker_coeff: whisker coefficient in box plot
    """
    Q1 = col.quantile(0.25)
    Q3 = col.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - whisker_coeff * IQR
    upper = Q3 + whisker_coeff * IQR
    return lower, upper


class BoxplotOutlierClipper(BaseEstimator, TransformerMixin):
    def __init__(self, whisker_coeff: float = 1.5):
        self.whisker = whisker_coeff
        self.lower = None
        self.upper = None

    def fit(self, X: pd.Series):
        self.lower, self.upper = find_boxplot_boundaries(X, self.whisker)
        return self

    def transform(self, X):
        return X.clip(self.lower, self.upper)
```

```python
# Vẽ lại biểu đồ sau khi loại bỏ outliers
clipped_total_rooms = BoxplotOutlierClipper().fit_transform(df_cmt["thank_count"])

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))
clipped_total_rooms.hist(bins=50, ax=axes[0])
clipped_total_rooms.to_frame().boxplot(ax=axes[1], vert=False);
```

```python
import pandas as pd

# Assuming 'sales_df' has columns 'Party', 'name', 'Date' (and potentially others)

# Remove duplicate records considering Party, product (name), and order date
unique_orders_df = sales_df.drop_duplicates(subset=['categories', 'name'])

# Find the number of orders and unique customers per product
most_popular_items_df = (
    unique_orders_df.groupby('name').agg(count_orders=('categories', 'nunique'), count_customers=('categories', 'nunique'))  # Corrected aggregation
)
most_popular_items_df.columns = ['No_of_Orders', 'No_of_Customers']  # Rename columns after aggregation
most_popular_items_df.reset_index(inplace=True)

# Calculate weighted number of orders (considering both frequency and reach)
M = most_popular_items_df['No_of_Customers'].max()  # Maximum number of customers
most_popular_items_df['Weighted_No_of_Orders'] = (
    most_popular_items_df['No_of_Orders'] * (most_popular_items_df['No_of_Customers'] / M)
)

# Rank products by weighted number of orders (descending)
most_popular_items_df['Popularity_Rank'] = (
    most_popular_items_df['Weighted_No_of_Orders'].rank(ascending=False).astype(int)
)

# Top 20 most popular items (considering weighted ranking)
top_20_popular = most_popular_items_df.sort_values(by='Popularity_Rank').head(20)
print(top_20_popular)
```

```python
# Merge Top Selling Items Rank and Popularity Rank dataframes
product_rankings_df = pd.merge(top_sell_items_df,most_popular_items_df,how='inner',on='name')

# Get only the Product, Price and Rank columns
product_rankings_df = product_rankings_df[['name','rating','Top_Sell_Rank','Popularity_Rank']]

# List the Product Rankings
#product_rankings_df.sort_values('Popularity_Rank',ascending=True).head(20)
```

```python
product_rankings_df.to_csv('Product-Rankings.csv',index=False)
```

```python
pickle.dump(product_rankings_df, open('prod_ranking_model.pkl','wb'))
```

PGM-3 - Customer-Product-Ranking

```python
import pandas as pd
import matplotlib.pyplot as plt
import pickle

# Import Feature Engineered Sales Transaction file
sales_df = pd.read_csv('df_merged.csv')
```

```python
# Find the no of units sold of each product by customer
top_sell_cust_items_df = sales_df.groupby(['categories','name']).agg({'quantity_sold':'sum'})

# Reset the index by converting the Party and Product into a column
top_sell_cust_items_df.reset_index(inplace=True)


# Rank the product by most Qty sold, at Customer level
categories_col = top_sell_cust_items_df['categories']
quantity_sold_col = top_sell_cust_items_df['quantity_sold'].astype(str)
top_sell_cust_items_df['Top_Sell_Rank'] = (categories_col + quantity_sold_col).rank(method='min',ascending=False).astype(int)


# List the top 20 items sold
#top_sell_cust_items_df.sort_values('Top_Sell_Rank',ascending=True).head(20)
```

```python
import pandas as pd

# Assuming 'sales_df' has columns 'Party', 'name', 'Date' (and potentially others)

# Remove duplicates considering Party, product (name), and order date
unique_orders_df = sales_df.drop_duplicates(subset=['categories', 'name'])

# Group by product category and name, calculate number of orders per group
freq_items_df = unique_orders_df.groupby(['categories', 'name']).size().to_frame(name='No_of_Orders')

# No need to reset index here (categories and name are already for grouping)

# Rank products within each customer category by number of orders (descending)
freq_items_df = freq_items_df.groupby('categories')['No_of_Orders'].rank(ascending=False).astype(int).to_frame('Popularity_Rank')
freq_items_df.reset_index(inplace=True)  # Reset index for ranking within customer

# Top 20 most popular items (considering correct ranking)
top_20_popular = freq_items_df.sort_values(by='Popularity_Rank').head(20)
```

```python
import pandas as pd
# Merge Top Selling Items Rank and Popularity Rank dataframes
cust_prod_rankings_df = pd.merge(top_sell_cust_items_df,freq_items_df,how='inner',on=['categories','name'])


# Merge the Unit Price (max price at product level)

# Find the unit price of each product (max of price considered, may required to be changed to median or mean)
items_price_df = sales_df.groupby(['name']).agg({'rating':'max'})

# Reset the index by converting the Party and Product into columns
items_price_df.reset_index(inplace=True)

# This ensures the same unit price is attached to the product purchased by different customers
cust_prod_rankings_df = pd.merge(cust_prod_rankings_df,items_price_df,how='left',on='name')


# Get only the Customer, Product, Price and Rank columns
cust_prod_rankings_df = cust_prod_rankings_df[['categories','name','rating','quantity_sold','Top_Sell_Rank','No_of_Orders','Popularity_Rank']]


# List the Product Rankings
#cust_prod_rankings_df.sort_values('Popularity_Rank',ascending=True).head(20)
```

```python
cust_prod_rankings_df.to_csv('Customer-Product-Rankings.csv',index=False)
```

```python
pickle.dump(cust_prod_rankings_df, open('cust_prod_ranking_model.pkl','wb'))
```

PGM-4 - Recommend_Products_to_Customer.

```python
import pandas as pd
import pickle

# Import Feature Engineered Sales Transaction file
sales_df = pd.read_csv('df_merged.csv')
```

```python
# Find the total qty purchased by each customer of each product
prod_cust_qty_df = sales_df.groupby(['name','categories']).agg({'quantity_sold':'sum'})

# Reset the index by converting the Party and Product into columns
prod_cust_qty_df.reset_index(inplace=True)


# Find the no of unique customers purchased each product
prod_cust_count_df = sales_df.groupby(['name']).agg({'categories':'nunique'})

# Set the customer count column
prod_cust_count_df.columns=['No_of_Customers']

# Reset the index by converting the Party and name into columns
prod_cust_count_df.reset_index(inplace=True)


# Merge the unique customer count and qty purchased of each name
prod_cust_df = pd.merge(prod_cust_qty_df,prod_cust_count_df,how='inner',on='name')


# Create a pivot table with all Customers on columns and names on rows, and Qty as values
prod_cust_pivot_df = prod_cust_df.pivot(index='name',columns='categories',values='quantity_sold').fillna(0)

# Find the correlation between every two customers and build a correlation matrix using corr() method
# Used Spearman method in identifying the correlation. Pearson was not providing better results and Kendall is taking a long time for execution.
cust_correlation_df = prod_cust_pivot_df.corr(method='spearman',min_periods=5)
#cust_correlation_df
```

```python
cust_correlation_df.to_csv('Customer-Customer-Correlation-Matrix.csv')
```

```python
pickle.dump(cust_correlation_df, open('cust_correlation_model.pkl','wb'))
```

```python
import pandas as pd
import os
from os  import getcwd
import pickle
from flask import Flask, render_template, request

app = Flask(__name__)

directory = getcwd()
```

```python
prod_ranking_model = pickle.load(open(os.path.join(directory,'prod_ranking_model.pkl'),'rb'))
cust_prod_ranking_model = pickle.load(open(os.path.join(directory,'cust_prod_ranking_model.pkl'),'rb'))
cust_correlation_model = pickle.load(open(os.path.join(directory,'cust_correlation_model.pkl'),'rb'))
prod_correlation_model = pickle.load(open(os.path.join(directory,'prod_correlation_model.pkl'),'rb'))
```

```python
# This function structures the HTML code for displaying the table on website
def html_code_table(prod_df,table_name,file_name,side):
    table_style = '<table style="border: 2px solid; float: ' + side + '; width: 40%;">'
    table_head = '<caption style="text-align: center; caption-side: top; font-size: 140%; font-weight: bold; color:black;"><strong>' + table_name + '</strong></caption>'
    table_head_row = '<tr><th>Product Name</th><th>Price (in Rs.)</th></tr>'

    html_code = table_style + table_head + table_head_row

    for i in range(len(prod_df.index)):
        row = '<tr><td>' + str(prod_df['name'][i]) + '</td><td>' + str(prod_df['rating'][i]) + '</td></tr>'
        html_code = html_code + row

    html_code = html_code + '</table>'

    file_path = os.path.join(directory,'templates/')

    hs = open(file_path + file_name + '.html', 'w')
    hs.write(html_code)

    #print(html_code)
```


Most Popular and Top Selling Products

```python
# This function calls the html_code_table function to create a .html file for Most Popular Products
def most_popular_table():
    most_popular_prods = prod_ranking_model.sort_values('Popularity_Rank',ascending=True)[['name','rating']].head(10).reset_index(drop=True)

    html_code_table(most_popular_prods,'Most Popular Products','mostpopulartable','left')

# This function calls the html_code_table function to create a .html file for Top Selling Products
def top_sell_table():
    top_sell_prods = prod_ranking_model.sort_values('Top_Sell_Rank',ascending=True)[['name','rating']].head(10).reset_index(drop=True)

    html_code_table(top_sell_prods,'Top Selling Products','topselltable','right')
```

Customer Frequently Purchased and Purchased the Most Products

```python
# This function calls the html_code_table function to create a .html file for Most Popular Products of a Customer
def cust_most_popular_table(cust_name):
    cust_most_popular_prods = cust_prod_ranking_model[cust_prod_ranking_model['categories'] == cust_name]
    cust_most_popular_prods = cust_most_popular_prods.sort_values('Popularity_Rank',ascending=True)[['name','rating']].head(10).reset_index(drop=True)

    html_code_table(cust_most_popular_prods,'Products you Frequently Purchased','custmostpopulartable','left')

# This function calls the html_code_table function to create a .html file for Top Selling Products of a Customer
def cust_top_sell_table(cust_name):
    cust_top_sell_prods = cust_prod_ranking_model[cust_prod_ranking_model['categories'] == cust_name]
    cust_top_sell_prods = cust_top_sell_prods.sort_values('Top_Sell_Rank',ascending=True)[['name','rating']].head(10).reset_index(drop=True)

    #print(cust_top_sell_prods)
    html_code_table(cust_top_sell_prods,'Products you Purchased the Most','custtopselltable','right')
```

Products Customer may Like

```python
# This function performs the below functionality for the input customer
# - get the list of customers with similar purchasing pattern and correlation coefficient
# - for each customer from the list,
#   - get the products purchased
#   - multiply the purchased qty with customer correlation coefficient
# - aggregate the qty_corr by product
# - ignore the products already purchased by the input customer
# - sort them by the qty_corr
# - calls the html_code_table function to create a .html file for top 10 products customer may like

def recommend_prod_cust(cust_name):
    similar_custs_corr = cust_correlation_model.loc[cust_name].sort_values(ascending=False)

    prod_by_similar_custs = pd.DataFrame()

    # get the products purchased by each customer and multiply with the customer correlation coefficient
    for i in range(len(similar_custs_corr)):
        if similar_custs_corr.index[i] != cust_name:
            cust_top_sell_prods = cust_prod_ranking_model[cust_prod_ranking_model['categories'] == similar_custs_corr.index[i]]
            cust_top_sell_prods = cust_top_sell_prods[['name','quantity_sold','rating']].reset_index(drop=True)
            cust_top_sell_prods['quantity_sold_Corr'] = cust_top_sell_prods['quantity_sold'] * similar_custs_corr.iloc[i]
            prod_by_similar_custs = pd.concat([cust_top_sell_prods,prod_by_similar_custs])

    # aggregate the Qty Correlation by Product
    prod_by_similar_custs = prod_by_similar_custs.groupby('name').agg({'quantity_sold_Corr':'sum','rating':'max'})
    prod_by_similar_custs.reset_index(inplace=True)
    #print(prod_by_similar_custs.head(20))

    # ignore the products already purchased by the input customer
    # merge prod_by_similar_custs and customer purchased products and drop the rows with No_of_orders being Not Null
    input_cust_top_sell_prods = cust_prod_ranking_model[cust_prod_ranking_model['categories'] == cust_name]
    df_merge = pd.merge(prod_by_similar_custs,input_cust_top_sell_prods[['name','No_of_Orders']],how='left',on='name')
    prod_recommend_to_cust = df_merge[df_merge['No_of_Orders'].isnull()]

    # sort the dataframe on Qty_Corr
    prod_recommend_to_cust = prod_recommend_to_cust.sort_values('quantity_sold_Corr',ascending=False)[['name','rating']].head(10).reset_index(drop=True)

    #print(prod_recommend_to_cust)

    html_code_table(prod_recommend_to_cust,'Products you may like','prodrecommendtable','center')
```

Similar Products to Display

```python
# This function performs the below functionality for the input product
# - get the list of products with similar purchasing pattern and correlation coefficient
# - get the price of each product from prod_ranking_model
# - get the price of input product and return to main
# - drop the product in view from the list
# - sort them by the correlation coefficient
# - calls the html_code_table function to create a .html file for top 10 products similar to the product in view

def similar_prods(prod_name):
    similar_prods_corr = prod_correlation_model.loc[prod_name].sort_values(ascending=False)

    similar_prods = pd.merge(similar_prods_corr,prod_ranking_model[['name','rating']],how='left',on='name')

    prod_price = similar_prods[similar_prods['name'] == prod_name]['rating'].values[0]

    input_prod_index = similar_prods[similar_prods['name'] == prod_name].index
    similar_prods.drop(index=input_prod_index,inplace=True)

    similar_prods = similar_prods[['name','rating']].head(10).reset_index(drop=True)

    #print(similar_prods)

    html_code_table(similar_prods,'Customers who purchased this product also purchased these','similarprodtable','left')

    return prod_price
```

```python
@app.route("/")
def home():
    most_popular_table()
    top_sell_table()

    return render_template('home.html')


@app.route("/login")
def login():
    most_popular_table()
    top_sell_table()

    cust_name = str(request.args.get('namecus')).upper()

    if cust_name in cust_prod_ranking_model['categories'].unique():
        cust_most_popular_table(cust_name)
        cust_top_sell_table(cust_name)
        recommend_prod_cust(cust_name)
        return render_template('cust_home.html',name=cust_name,new='n')
    else:
        return render_template('cust_home.html',name=cust_name,new='y')


@app.route("/view")
def view():
    prod_name = str(request.args.get('prod')).upper()

    if prod_name in prod_ranking_model['name'].unique():
        prod_price = similar_prods(prod_name)
        return render_template('prod_view.html',prod=prod_name,price=prod_price,exists='y')
    else:
        return render_template('prod_view.html',prod=prod_name,exists='n')


if __name__ == "__main__":
    app.run(debug=True)
```
